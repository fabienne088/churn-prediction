```python
def explore_dataframe(df):
    """ Explore a pandas DataFrame.

    Parameters:
    df (DataFrame): Input DataFrame to explore.
    """

    # Display the shape of the DataFrame
    print(f"DataFrame shape: {df.shape}")

    # Display the first few rows of the DataFrame
    print(f"First few rows of the DataFrame:")
    display(df.head())

    # Display the summary statistics of the DataFrame
    print(f"Summary statistics of the DataFrame:")
    description = df.describe()
    display(description.round(2))

    # Display the information about the DataFrame
    print(f"Information about the DataFrame:")
    display(df.info())

    # Check for missing values in the DataFrame
    print(f"Total of Missing values for each column in the DataFrame:")
    print(df.isnull().sum())

    # Checking for duplicated values
    print(f"\n Number of duplicates in the DataFrame = {df.duplicated().sum()}")
```

```python
def plot_bar_pie(df, feature):
    """Plot bar and pie charts for a given feature"""
    # Calculate counts of each feature
    counts = df[feature].value_counts()
    # Create figure and axes
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Bar plot
    sns.countplot(data=df, x=feature, hue= 'Attrition_Flag', order=counts.index, stat='percent',palette='viridis', ax=axs[0])
    axs[0].set_title(f'{feature} Distribution (Bar)')
    axs[0].set_xlabel(feature)
    axs[0].set_ylabel('Percent (%)')
    axs[0].tick_params(axis='x', rotation=45)
    axs[0].bar_label(axs[0].containers[0], fontsize=10, fmt='%.1f%%')

    # Pie chart
    size = 0.5
    axs[1].pie(counts, labels=counts.index, 
            autopct='%1.1f%%', pctdistance=0.7,
            startangle=30, explode=[0.02] * len(counts),
            wedgeprops=dict(width=size, edgecolor='w'))
    axs[1].set_title(f'{feature} Distribution (Pie)')

    # Adjust layout to prevent overlap
    plt.tight_layout()

    # Show plot
    plt.show()
```

```python
def handle_outliers(df, columns_to_remove_outliers, zscore_threshold=3):
    """
    Handle outliers in specified columns of a DataFrame.
    Parameters:
        df (DataFrame): Input DataFrame.
        columns_to_remove_outliers (list): List of column names to remove outliers.
        zscore_threshold (float): Z-score threshold for identifying outliers. Default is 3.
    Returns:
        DataFrame: DataFrame with outliers removed.
    """
    # Create a copy of the DataFrame to avoid modifying the original DataFrame
    df_outliers = df.copy()

    # Iterate through each column in columns_to_remove_outliers
    for column_name in columns_to_remove_outliers:
        # Convert the column to a numeric data type, ignoring errors
        df_outliers[column_name] = pd.to_numeric(df_outliers[column_name], errors='coerce')

        # Calculate Z-scores for the specified column
        z_scores = (df_outliers[column_name] - df_outliers[column_name].mean()) / df_outliers[column_name].std()

        # Create a mask to identify outliers
        outlier_mask = np.abs(z_scores) > zscore_threshold

        # Print the columns for which outliers are being removed
        print("Removing outliers for column: " + column_name)

        # Count missing values
        missing_values_count = df_outliers[column_name].isna().sum()
        print("Number of missing values in " + column_name + ": " + str(missing_values_count))

    return df_outliers

# Call the modified function
columns_to_remove_outliers = churners_clean.select_dtypes(exclude = 'object').columns
house_filtered = handle_outliers(churners_clean, columns_to_remove_outliers)

churners_clean.shape
```

```python
numeric = churners_clean.select_dtypes(exclude='object').columns.tolist()

def remove_outliers(df, num_cols, threshold=3):
    df_outliers = df.copy()
    # Calculate Z-scores for numerical columns
    z_scores = stats.zscore(df_outliers[num_cols])
    
    # Find outliers indices based on the threshold
    outliers_indices = (np.abs(z_scores) > threshold).any(axis=1)

    # Get the columns with outliers
    outlier_columns = [col for col, has_outliers in zip(num_cols, 
                        (np.abs(z_scores) > threshold).any(axis=0)) if has_outliers]
    
    # Remove outliers
    cleaned_df = df[~outliers_indices]
    
    # Count the values that are removed
    removed_count = outliers_indices.sum()

    return cleaned_df, outlier_columns, removed_count

# Call the function to remove outliers from numerical columns
cleaned_df, outlier_columns, removed_count = remove_outliers(churners_clean, numeric)

print('Columns with outliers:', outlier_columns)
print('Count of removed values:', removed_count)
print(cleaned_df.shape)
```
